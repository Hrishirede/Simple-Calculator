# Business Problem
In today"s buisy and fast life, people need simple to use and quick calculator. Its very difficult to use a very complex calculator for quick calculations. For simple numeric calculations people require simple and easy to use calculator for better productivity.


# Calculator by using Tkinter

This Simple Calculator is designed considering the basic requirement of a calculator by today"s professionals. This Calculator will definately improve productivity and save your time.

```python
from tkinter import *

root=Tk()
root.title('Calculator')
root.geometry('570x600+100+200')
root.resizable(0,0)
root.configure(background='#17161b')

equation=""

def User_input(value):
    global equation
    equation=equation+value
    result_window.config(text=equation)
    
def Clear():
    global equation
    equation=""
    result_window.config(text=equation)
    
def Equals():
    global equation
    result=""
    if equation!="":
        try:
            result=eval(equation)
        except:
            result="Error"         
    result_window.config(text=result)

def backspace():
    global equation
    equation=equation.rstrip(equation[-1])
    result_window.config(text=equation)
         

result_window =Label(root,width=25,height=2,text='',font=('verdana', 30))
result_window.pack()

Button(root,text='AC',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#3697f5',fg='#fff',command=lambda:Clear()).place(x=10,y=110)
Button(root,text='C',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#808088',fg='#fff',command=lambda:backspace()).place(x=150,y=110)
Button(root,text='.',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('.')).place(x=290,y=110)
Button(root,text='+',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('+')).place(x=430,y=110)

Button(root,text='7',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('7')).place(x=10,y=210)
Button(root,text='8',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('8')).place(x=150,y=210)
Button(root,text='9',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('9')).place(x=290,y=210)
Button(root,text='-',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('-')).place(x=430,y=210)

Button(root,text='4',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('4')).place(x=10,y=310)
Button(root,text='5',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('5')).place(x=150,y=310)
Button(root,text='6',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('6')).place(x=290,y=310)
Button(root,text='*',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('*')).place(x=430,y=310)

Button(root,text='1',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('1')).place(x=10,y=410)
Button(root,text='2',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('2')).place(x=150,y=410)
Button(root,text='3',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('3')).place(x=290,y=410)
Button(root,text='/',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('/')).place(x=430,y=410)

Button(root,text='(',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('(')).place(x=10,y=510)
Button(root,text='0',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input('0')).place(x=150,y=510)
Button(root,text=')',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#2a2d36',fg='#fff',command=lambda:User_input(')')).place(x=290,y=510)
Button(root,text='=',width=5,height=1,font=('arial',30,'bold'),bd=1, bg='#fe9037',fg='#fff',command=lambda:Equals()).place(x=430,y=510)

root.mainloop()
```
